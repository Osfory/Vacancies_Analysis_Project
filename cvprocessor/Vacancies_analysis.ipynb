В целом, кусок собранного датасета выглядит вполне неплохо. Теперь надо бы привести его в адекватный вид. Зададим более осмысленные названия для столбцов. А также нормальные индексы для строк.

```python
# Взглянем на количество вакансий
print(vacancies.shape)
```

Почти 300 вакансий! В целом, этого должно быть достаточно для анализа.

```python
vacancies.columns = ['vacancy_title', 'company_name', 'location', 'required_experience', 'employment_mode', 'vacancy_salary',
              'published_date', 'vacancy_description', 'vacancy_tags']
```

```python
vacancies.index = [idx for idx in range(1,vacancies.shape[0]+1)]
```

```python
vacancies.head(5)
```

Уже лучше. Но всё же есть ещё одна проблема - описание вакансий и тэги на данный момент являются цельными строками, без нормальной разметки. Исправим.

```python
vacancies.vacancy_tags = vacancies.vacancy_tags.str.split(pat="\n")
vacancies.vacancy_description = vacancies.vacancy_description.str.split(pat="\n")
vacancies.head()
```

Вот теперь намного лучше! С этим уже можно работать.
